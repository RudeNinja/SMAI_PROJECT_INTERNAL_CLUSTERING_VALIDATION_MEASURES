```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_gaussian_quantiles

import matplotlib.pyplot as plt
import csv
```

```python
plt.title("Multi-class, two informative features, one cluster", fontsize="small")
X1, Y1 = make_classification( n_samples=250,
    n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3
)
plt.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1, s=25, edgecolor="k")
```

```python
plt.title("Well Separated data", fontsize="small")
X1, Y1 = make_blobs(n_samples=250,n_features=2, centers=5,cluster_std=0.5,random_state=20)
plt.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1, s=25, edgecolor="k")
```

```python
print(X1.shape)

filename = "Wellseparated.csv"
with open(filename, 'w') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile) 
        
    # writing the fields 
    
        
    # writing the data rows 
    csvwriter.writerows(X1)
```

```python
arr = np.loadtxt(filename,delimiter=',')

print(arr)

plt.plot(arr[:,0],arr[:,1],'.')
plt.title("Well separated Data")
```
